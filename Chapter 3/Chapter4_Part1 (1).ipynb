Two interfaces for the price of One

```python
plt.figure() # create a plot figure

# create 1st of true panels
plt.subplot(2, 1, 1) # (row, cols, panel number)
plt.plot(x, np.sin(x))

# create 2nd panel
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

Adjusting plot: line colors and styles

```python
# specify color by name
plt.plot(x, np.sin(x-0), color='blue')
# short color code (rgbcmyk)
plt.plot(x, np.sin(x-1), color='g')
# grayscale between 0 and 1
plt.plot(x, np.sin(x-2), color='0.75')
# hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x-3), color='#FFDD44')
# RGB tuple, values 0 and 1
plt.plot(x, np.sin(x-4), color=(1.0, 0.2, 0.3))
# all HTML color names supported
plt.plot(x, np.sin(x-5), color='chartreuse')
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short
plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

```python
# extremely terse
plt.plot(x, x+0, '-g')
plt.plot(x, x+1, '--b')
plt.plot(x, x+2, '-.r')
plt.plot(x, x+3, ':c')
```

Adjusting the Plot: Axes Limits

```python
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
# if want to reverse
# reverse the order
plt.plot(x, np.sin(x))

plt.xlim(11, -1)
plt.ylim(1.5, -1.5)
```

```python
# plt.axis() sets x,y limits
# with 1 single call
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5])
plt.axis('tight')
```

```python
# plt.axis() equals aspect ratio
plt.plot(x, np.sin(x))
plt.axis('equal')
```

Labeling plot

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend()
```

## Visualizing Errors

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy*np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k')
```

```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0)
```

Continuous Error

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C

# define models and draw some data
model = lambda x: x * np.sin(x)
xdata = np.array([1,3,5,6,8])
ydata = model(xdata)

# compute gaussian process fit
kernel = C(1.0, (1e-3, 1e3)) * RBF(10.0, (1e-2, 1e2))
gp = GaussianProcessRegressor(kernel=kernel, alpha=1e-10, normalize_y=True)
gp.fit(xdata[:, np.newaxis], ydata)
```

```python
xfit = np.linspace(0, 10, 1000)
yfit, ystd = gp.predict(xfit[:, np.newaxis], return_std=True)
dyfit = 2 * ystd # 2*sigma ~ 95% confidence region
```

```python
# visualize result
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')

plt.fill_between(xfit, yfit - dyfit, yfit + dyfit,
                 color='gray', alpha=0.2)
plt.xlim(0, 10)
```

## Density and Contour Plots

```python
# Visualizing a 3-dimensional function
def f(x, y):
  return np.sin(x) ** 10 + np.cos(10 + y*x)*np.cos(x)
```

```python
# build 2-dimensional grids
# from 1-dimensional arrays
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors='black')
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy') # RedGray
```

```python
plt.cm.Blues
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar()
```

```python
plt.imshow(Z, extent=[0,5,0,5], origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis('image')
```

```python
contour = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contour, inline=True, fontsize=12)

plt.imshow(Z, extent=[0,5,0,5], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar()
```

## Histograms, Binnings, and Density

```python
data = np.random.randn(1000)
plt.hist(data)
```

```python
# customized histogram
plt.hist(data, bins=30, density=True, alpha=0.5, histtype='stepfilled',
         color='steelblue', edgecolor='none')
```

```python
# histype = 'stepfilled': useful when
# comparing histograms of several distributions
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```

```python
# just count the numbers of points in
# a given bin
# but not display
counts, bin_edges = np.histogram(data, bins=5)
counts
```

```python
# 2-dimensional histograms and binnings
mean = [0, 0]
cov = [[1,1], [1,2]]
x, y = np.random.multivariate_normal(mean, cov, 1000).T

x,y
```

```python
plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
counts, xedges, yedges = np.histogram2d(x, y, bins=30)
counts, xedges, yedges
```

plt.hexbin: hexagonal binnings

```python
plt.hexbin(x, y, gridsize= 30, cmap='Blues')
cb = plt.colorbar(label='count in bin')
```

kernel density estimation

```python
from scipy.stats import gaussian_kde

# fit an array of size [Ndim, Nsamples]
data = np.vstack([x, y])
kde = gaussian_kde(data)

# evaluate on a regular grid
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))
```
