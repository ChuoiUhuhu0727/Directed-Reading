## Exercise 4: Visualize Errors

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C

# Simulated e-commerce & fintech sales dataset
data = {
    'day': np.arange(1, 11),
    'sales_amount': [1200,1300,1100,1500,1600,1700,1400,1350,1250,1450],
    'expected_sales': [1150,1250,1120,1400,1550,1620,1350,1300,1200,1400],
    'std_sales': [100,120,90,130,150,160,110,80,100,110],
    'ad_spend': [200,210,180,220,250,260,210,205,195,230],
    'discount_rate': [0.05,0.10,0.07,0.12,0.15,0.20,0.08,0.05,0.06,0.10]
}
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# Error bar plot:
# sales_amount by day, error=std_sales
plt.figure(figsize=(10,6))
plt.errorbar(df['day'], df['sales_amount'], yerr=df['std_sales'], fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0)
plt.xlabel("Day")
plt.ylabel("Sales Amount")
plt.title("Daily Sales with Error Bars")
plt.show()
```

```python
# Gaussian Process Regression:
# sales_amount vs ad_spend
X = df['ad_spend'].values
y = df['sales_amount'].values
```

```python
kernel = C(1.0, (1e-3, 1e3)) * RBF(10.0, (1e-2, 1e2))
gp = GaussianProcessRegressor(kernel=kernel, alpha=df['std_sales'].mean(), normalize_y=True)
gp.fit(X[:, np.newaxis], y)

Xfit = np.linspace(X.min(), X.max(), 100)
yfit, ystd = gp.predict(Xfit[:, np.newaxis], return_std=True)
dyfit = 2 * ystd # 95% confidence region
```

```python
plt.figure(figsize=(10,6))
plt.plot(X, y, 'or', label='Observed Sales')
plt.plot(Xfit, yfit, '-', color='gray', label='GP Prediction')
plt.fill_between(Xfit, yfit - dyfit, yfit + dyfit, color='gray', alpha=0.2, label='95% CI')
plt.xlabel("Ad Spend")
plt.ylabel("Sales Amount")
plt.title("GP Regression: Sales vs Ad Spend with 95% Confidence Interval")
plt.legend()
plt.show()
```

## Exercise 5: Density and Contour Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Simulated ecommerce & fintech profit optimization dataset
data = {
    'price': np.repeat([10, 20, 30, 40, 50], 5),
    'discount_rate': np.tile([0.00, 0.05, 0.10, 0.15, 0.20], 5),
    'profit': [
        120, 115, 110, 105, 100,
        200, 195, 190, 185, 180,
        270, 265, 260, 255, 250,
        320, 315, 310, 305, 300,
        350, 345, 340, 335, 330
    ]
}
df = pd.DataFrame(data)
df
```

```python
# Prepare grids
x = np.unique(df['price'])
y = np.unique(df['discount_rate'])
X, Y = np.meshgrid(x, y)
Z = np.zeros_like(X)

for i in range(len(x)):
    for j in range(len(y)):
        Z[j, i] = df[(df['price'] == x[i]) & (df['discount_rate'] == y[j])]['profit'].values[0]
```

```python
# Contour plot
plt.figure(figsize=(10, 6))
contour = plt.contour(X, Y, Z, colors='black')
plt.clabel(contour, inline=True, fontsize=12)
plt.contour(X, Y, Z, 10, cmap='RdGy')
plt.contourf(X, Y, Z, 10, cmap='RdGy')
plt.colorbar(label='Profit')
plt.xlabel('Price')
plt.ylabel('Discount Rate')
plt.title('Profit Surface: Price vs Discount Rate (Contour)')
plt.show()
```

```python
# Imshow overlay
plt.figure(figsize=(10, 6))
plt.imshow(Z, extent=[x.min(), x.max(), y.min(), y.max()], origin='lower', cmap='RdGy', alpha=0.5, aspect='auto')
plt.colorbar(label='Profit')
plt.xlabel('Price')
plt.ylabel('Discount Rate')
plt.title('Profit Surface: Price vs Discount Rate (Imshow Overlay)')
plt.show()
```

## Exercise 6: Histograms, Binnings, and Density

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Simulate ecommerce/fintech data
np.random.seed(42)
transaction_amount = np.random.normal(100, 50, 1000)  # mean=100, std=50
customer_age = np.random.normal(35, 10, 1000)
product_price = np.random.normal(40, 20, 1000)
days_since_last_purchase = np.random.exponential(10, 1000)
```

```python
# 1D histogram of transaction amounts
plt.figure(figsize=(8,4))
plt.hist(transaction_amount)
plt.title('Histogram of Transaction Amounts')
plt.xlabel('Transaction Amount')
plt.ylabel('Count')
plt.show()
```

```python
# Customized histogram
plt.figure(figsize=(8,4))
plt.hist(transaction_amount, bins=30, density=True, alpha=0.5, histtype='stepfilled',
         color='steelblue', edgecolor='none')
plt.title('Histogram of Transaction Amounts')
plt.xlabel('Transaction Amount')
plt.ylabel('Density')
plt.show()
```

```python
# Compare distributions: segments
# (regular, premium, new customers)
x1 = np.random.normal(80, 30, 1000)    # Regular
x2 = np.random.normal(150, 60, 1000)   # Premium
x3 = np.random.normal(50, 20, 1000)    # New

kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=40)
plt.figure(figsize=(8,4))
plt.hist(x1, **kwargs, label="Regular")
plt.hist(x2, **kwargs, label="Premium")
plt.hist(x3, **kwargs, label="New")
plt.legend()
plt.title('Segmented Transaction Amount Distributions')
plt.xlabel('Transaction Amount')
plt.ylabel('Density')
plt.show()
```

```python
# 2D histogram: Amount vs Customer Age
plt.figure(figsize=(8,6))
plt.hist2d(transaction_amount, customer_age, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('Counts in bin')
plt.xlabel('Transaction Amount')
plt.ylabel('Customer Age')
plt.title('2D Histogram: Transaction Amount vs Customer Age')
```

```python
# Hexbin plot
plt.figure(figsize=(8,6))
plt.hexbin(transaction_amount, customer_age, gridsize=30, cmap='Blues')
cb = plt.colorbar(label='count in bin')
plt.xlabel('Transaction Amount')
plt.ylabel('Customer Age')
plt.title('Hexbin: Transaction Amount vs Customer Age')
```

```python
# Gaussian KDE density estimation
data = np.vstack([transaction_amount, customer_age])
kde = gaussian_kde(data)

xgrid = np.linspace(transaction_amount.min(), transaction_amount.max(), 40)
ygrid = np.linspace(customer_age.min(), customer_age.max(), 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

plt.figure(figsize=(8,6))
plt.imshow(Z.reshape(Xgrid.shape),
           origin='lower', aspect='auto',
           extent=[xgrid.min(), xgrid.max(), ygrid.min(), ygrid.max()],
           cmap='Blues')
plt.colorbar(label="Density")
plt.xlabel('Transaction Amount')
plt.ylabel('Customer Age')
plt.title('KDE: Transaction Amount vs Customer Age')
plt.show()
```
